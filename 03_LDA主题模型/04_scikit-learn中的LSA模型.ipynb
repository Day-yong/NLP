```python
# 加载数据
with open('./data/1.txt', 'r', encoding='utf-8') as f1:
    res1 = f1.read()
with open('./data/2.txt', 'r', encoding='utf-8') as f2:
    res2 = f2.read()
with open('./data/3.txt', 'r', encoding='utf-8') as f3:
    res3 = f3.read()
contents = [res1, res2, res3]
print(contents)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import TruncatedSVD
```

```python
# 将文本数据转换为向量的形式(词袋法) 要求给定的数据中单词是以空格隔开的
cntVector = CountVectorizer()
cnt = cntVector.fit_transform(contents)
print(cnt.shape)
print(cnt.toarray())
```

```python
# LSA模型（即SVD）
svd = TruncatedSVD(n_components=2)
docs = svd.fit_transform(cnt)
```

```python
# 打印文档和主题之间的相关性
print(docs)
# 打印主题和词之间的相关性
print(svd.components_)
```
